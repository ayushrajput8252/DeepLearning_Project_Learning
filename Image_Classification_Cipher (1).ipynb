```python
import keras
from keras.datasets import cifar10
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, BatchNormalization
from keras.layers import Conv2D, MaxPooling2D, Flatten
from sklearn.metrics import confusion_matrix
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
x_train.shape
```

```python
#ONE HOT ENCODING
```

```python
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)
```

```python
#FEATURE ENGINNERING
```

```python
x_train.dtype
```

```python
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train = x_train/255
x_test = x_test/255
```

```python
#MODEL ARCTITECTURE
```

```python
model = Sequential()
#Layer 1
model.add(Conv2D(32, (3,3), kernel_initializer='he_normal', activation='relu', input_shape=x_train.shape[1:]))
model.add(BatchNormalization())
#Layer 2
model.add(Conv2D(32, (3,3), kernel_initializer='he_normal', activation='relu'))
model.add(BatchNormalization())
#Layer 3
model.add(Conv2D(32, (3,3), kernel_initializer='he_normal', activation='relu'))
model.add(BatchNormalization())
#Layer 4
model.add(Conv2D(48, (3,3), kernel_initializer='he_normal', activation='relu'))
model.add(BatchNormalization())
#Layer 5
model.add(Conv2D(48, (3,3), kernel_initializer='he_normal', activation='relu'))
model.add(BatchNormalization())
#Layer 6
model.add(Conv2D(64, (3,3), kernel_initializer='he_normal', activation='relu'))
model.add(BatchNormalization())
#Layer 7
model.add(Conv2D(128, (3,3), kernel_initializer='he_normal', activation='relu'))
model.add(BatchNormalization())

model.add(MaxPooling2D())
model.add(Dropout(0.2))
model.add(Flatten()) # Add Flatten layer here
model.add(Dense(10, activation='softmax')) # Add Dense output layer with 10 units for CIFAR-10 classes
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, batch_size=500, epochs=100, validation_data=(x_test, y_test))
```
